```python
# 初始化

import numpy as np
import pandas as pd
import matplotlib as mpl
from matplotlib import pyplot as plt

mpl.rcParams['font.family'] = 'SimHei'

data = pd.read_csv('../data/prodData_YT_Sorted.csv')
data.info()
```

```python
averProd = list(data['AverProd'])
averProd.sort()
print(averProd[:3])
print(averProd[-3:])
```

```python
# 根据城市分别获取数据并绘制

plt.xlim(1992, 2021)  
plt.ylim(0, 70)
groupData = data.groupby('Name')
for i in list(groupData):
    cityData = i[1]
    x = list(cityData['Time'])
    y = list(cityData['AverProd']) 
    plt.plot(x, y, label = i[0])
# (太多了不推荐) 去掉下方代码注释以查看图例
# plt.legend()
```

```python
# 数据有几个比较怪的，分城市绘制一下试试

count = len(groupData)
for i in range(count):
    fig      = plt.figure(i)
    cityName = list(groupData)[i][0]
    cityData = list(groupData)[i][1]
    x = list(cityData['Time'])
    y = list(cityData['AverProd']) 
    plt.xlim(1992, 2021)
    plt.ylim(0, 70)
    plt.plot(x, y)
    plt.title(cityName)
```
